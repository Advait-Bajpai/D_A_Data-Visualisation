```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

## Scatterplot

```python
sns.scatterplot(x = "age", y = "fare", data = df, hue = "survived")
```

```python
df2 = sns.load_dataset("tips")
df2.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df2)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df2, hue= "sex")
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df2, hue= "sex", style = "smoker")
```

## Jointplot

```python
sns.jointplot(x= "total_bill", y= "tip",hue = "smoker" ,data = df2)
```

```python
sns.jointplot(x= "total_bill", y= "tip",data = df2, kind ="reg")
```
