```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.barplot(x = "class", y = "age", data = df, estimator='mean')
```

```python
sns.barplot(x = "class", y = "fare", data = df, estimator='sum')
```

```python
sns.barplot(x = "class", y = "fare", hue = "sex", data = df, estimator='sum')
```

```python
sns.boxplot(x = "class", y = "age", data = df, )
```

```python
sns.boxplot(x = "class", y = "age", data = df, hue = "survived" )
```

```python
sns.violinplot(x = "class", y = "age", data = df, )
```

```python
sns.distplot(df[df["class"] == "Second"]["age"])# It is to plot the probability distribution
```
